```python
from itertools import combinations

# Given set
S = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

# Generate all combinations of size 5
subset_size = 5
zero_sum_subsets = [subset for subset in combinations(S, subset_size) if sum(subset) == 0]

# Print the results
for subset in zero_sum_subsets:
    print(subset)
```

```python
from itertools import combinations

# Given set
S = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

# Collect zero-sum subsets of size 3 to 6
zero_sum_subsets = []

for size in range(3, 7):  # sizes 3 to 6
    for subset in combinations(S, size):
        if sum(subset) == 0:
            zero_sum_subsets.append(subset)

# Print the results
for subset in zero_sum_subsets:
    print(subset)
```
